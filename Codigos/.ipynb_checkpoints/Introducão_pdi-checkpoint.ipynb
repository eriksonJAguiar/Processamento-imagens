**Bibliotecas utilizadas**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
path_img = "cat.jpeg"
```

**Ler imagens**

```python
img = cv2.imread(path_img)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

**Visualização do shape da imagem**

```python
print("Imagem Colorida Shape: {}".format(img_rgb.shape))
print("Imagem Tons de cinza Shape: {}".format(img_gray.shape))
```

## **Plotar  a imagem**

**Colorida**

```python
plt.imshow(img_rgb)
plt.axis("off")
plt.show()
```

```python
plt.imshow(img_rgb[:,:,0])
plt.axis("off")
plt.show()
```

```python
plt.imshow(img_rgb[:,:,1])
plt.axis("off")
plt.show()
```

```python
plt.imshow(img_rgb[:,:,2])
plt.axis("off")
plt.show()
```

**Tons de cinza**

```python
plt.imshow(img_gray, cmap="gray")
plt.axis("off")
plt.show()
```
