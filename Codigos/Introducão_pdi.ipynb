**Bibliotecas utilizadas**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
path_img = "cat.jpeg"
```

**Ler imagens**

```python
def input_image(img_path, is_gray=False):
   img = cv2.imread(img_path)
   img_new = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) if is_gray else cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
   
   return img_new
```

```python
img = cv2.imread(path_img)
#img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
#img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
img_rgb = input_image(path_img)
img_gray = input_image(path_img, is_gray=True)
```

**Visualização do shape da imagem**

```python
print("Imagem Colorida Shape: {}".format(img_rgb.shape))
print("Imagem Tons de cinza Shape: {}".format(img_gray.shape))
```
